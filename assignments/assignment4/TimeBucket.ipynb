Test how to deal with hour time buckets

```python
import time

first = True
last = None
with open("WC2015_1.csv","r") as f:
    for line in f:
        if not first:
            t=line.split("|")[5].lower()
            tm = time.gmtime(int(t))
            tmt = (tm.tm_mon, tm.tm_mday, tm.tm_hour)
            if last:
                if tmt != last:
                    print "New hour " + str(tm.tm_hour)
            else:
                print "New hour " + str(tm.tm_hour)
            last = tmt
        else:
            first = False
        
```
